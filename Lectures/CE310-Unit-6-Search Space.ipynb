Let's compute the number N(d) of different trees of maximum depth d, which represents the search space explored by the GP.

```python
def N(d,C):
    # This function computers the number of programs with up to d levels 
    # (single terminals having d=0) that can be constructed with a given 
    # primitive set. All we need to know about the primitive set is the 
    # arity (number of arguments) of its elements. These are provided as 
    # a list in C. Note that also terminals (arity-zero elements) need 
    # to be listed.
    if d == 0:
        return len([p for p in C if p == 0])
    else:
        sum = 0
        ndm1 = N(d-1,C)
        for p in C:
            sum += ndm1 ** p
        return sum

# AND OR NOR NAND x1 x2 x3
#print(N(0,[2, 2, 2, 2, 0, 0, 0]))
#print(N(1,[2, 2, 2, 2, 0, 0, 0]))
#print(N(2,[2, 2, 2, 2, 0, 0, 0]))
#print(N(3,[2, 2, 2, 2, 0, 0, 0]))
#print(N(6,[2, 2, 2, 2, 0, 0, 0]))
#print(N(17,[2, 2, 2, 2, 0, 0, 0]))


# standard primitive set used in the CE310 mini project
# +, -, *. %, sin, cos, +1, -1, rand
print(N(0,[2, 2, 2, 2, 1,1,0, 0, 0]))
print(N(1,[2, 2, 2, 2, 1,1,0, 0, 0]))
print(N(2,[2, 2, 2, 2, 1,1,0, 0, 0]))
print(N(3,[2, 2, 2, 2, 1,1, 0, 0, 0]))
print(N(4,[2, 2, 2, 2, 1,10, 0, 0]))
print(N(5,[2, 2, 2, 2, 1,1,0, 0, 0]))
#print(N(6,[2, 2, 2, 2, 1,1,0, 0, 0]))
#print(N(7,[2, 2, 2, 2, 1,1,0, 0, 0]))

# How many years of calculation?
tests_per_second = 1e10
N(5,[2, 2, 2, 2, 0, 0, 0])/(tests_per_second*60*60*24*365)
```
